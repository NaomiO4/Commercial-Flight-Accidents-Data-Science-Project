```python
def drop_cols(commercialflights):
    """
    A function that cleans dataset 1 of columns and null values in the ProbableCause column.
    """
    commercialflights.shape
    commercialflights= commercialflights[['EventDate','HighestInjuryLevel', 'FatalInjuryCount', 
                                          'SeriousInjuryCount', 'MinorInjuryCount', 'ProbableCause',
                                          'Make', 'Model','AirCraftDamage', 
                                          'WeatherCondition', 'Operator']]
    commercialflights = commercialflights.dropna(axis=0, subset="ProbableCause")
    commercialflights.shape
    return commercialflights
```

One important step to prepare our data for analysis is that we need to categorize the description of the probable cause for each crash into our categories of interest: weather, human error, and mechanical error. To do so, we first prepare the text by converting it into a format that can be cleaned.

```python
def clean_text(text):
    """
    Cleans the text by removing punctuation and splitting it into words. Converts everything to lowercase.
    """
    punctuation_chars = ".,!?;:'\"()[]{}<>-/_&"
    
    # Replace punctuation with spaces
    for char in punctuation_chars:
        text = text.replace(char, ' ')
    
    # Convert text to lowercase and split into words
    words = text.lower().split()
    return words
```

Next, we identify words that appear very often in the probable causes that need to be removed because they are not informative for our purposes of categorizing the crashes by their cause. 

```python
# List of common words to exclude (stop words)
# Stop words determined first via common word articles
# Then running the functions, manually adding more words that are not causes, e.g. 'during' and 'resulted'

stop_words = set([
    # the following are filler words
    'the', 'a', 'an', 'and', 'in', 'on', 'or', 'at', 'of', 'to', 'for', 'with', 'about', 'as', 'by', 'this', 'that', 's', 'was', 'which', 'during', 'resulted', 'from', 'were', 'due', 'his',
    'subsequent', 'clear', 'between', 'when', 'use', 'their', 'caused', 'being', 'after', 'into',
    # the words below are usually found in explanations but they are not the causes
    'contributing', 'airplane', 'flight', 'accident', 'aircraft', 'not', 'factor', 'factors', 'resulting', 'failure', 'inadequate', 'inadvertent', 'encounter', 'landing',
    'incident', 'control', 'improper', 'right', 'left', 'company', 'system', 'while', 'lack', 'maintain', 'approach', 'takeoff', 'main', 'loss', 'fatigue', 'follow',
    'runway', 'procedures', 'injury', 'ground', 'clearance', 'fatigue', 'first', 'tug', 'inspection', 'manufacturer', 'collision', '1', 'undetermined', 'area', 'serious',
    '2', 'cabin', 'descent', 'door', 'local', 'airport', 'adequate', 'excessive', 'visual', 'high', 'information', 'reasons', 'proper', 'damage',
    # this word is a type of turbulence
    'convective'
])
```

Now we can remove the unneeded words to reduce each cause to their key words.

```python
def process_row(text, stop_words):
    """
    Processes a single row of text by cleaning and filtering out stop words. Returns a list of words.
    """
    # Clean the text and split it into words
    words = clean_text(text)
    
    # Filter out stop words
    filtered_words = [word for word in words if word not in stop_words]
    
    return filtered_words
```

```python
commercialflights['ProbableCause'] = commercialflights['ProbableCause'].apply(lambda x: ' '.join(process_row(str(x), stop_words)))
```

Now that we are left with words that indicate how each crash occurred, let's take a look at the most common words that occur in the crash descriptions.

```python
def count_top_words(csv_file, column_name, top_n):
    """
    Reads CSV file row by row, processes the specified column to count word frequencies
    excluding stop words, and returns the top N most common words.
    The results will be used in categorizing incident causes for data analysis.
    """
    # Initialize a Counter to count word frequencies
    word_counts = Counter()
    
    # Clean dataset using previously-defined drop_cols function
    cleaned = drop_cols(commercialflights)
    
    for text in commercialflights[column_name].dropna():
        filtered_words = process_row(str(text), stop_words)
        word_counts.update(filtered_words)
    
    # Get the top N most common words
    most_common_words = word_counts.most_common(top_n)
    
    return most_common_words
```

```python
column_name = 'ProbableCause'
top_n = 30  # Number of top words we want to see

# Get top N most common words
most_repeated = count_top_words(commercialflights, column_name, top_n)

# Display result
most_repeated
```

```python
overallflights
```

This dataset includes flights that are not commercial, so we want to remove any flight that is military, mail delivery, or air taxi.

```python
#Function: drop rows with military, mail, air taxi values in the operator column.

def drop_specific_operators(df):
    keywords = ["military", "mail", "air taxi"]
    df = df.dropna(subset=["Operator"])  # Ensure 'Operator' column has no NaN values
    pattern = '|'.join(keywords)  # Create a regex pattern from keywords
    filtered_df = df[~df["Operator"].str.contains(pattern, case=False, na=False)] #Case-insensitive filtering
    return filtered_df

#apply to dataset 
overallflights = drop_specific_operators(overallflights)
```

```python
overallflights
```

The dataset still contains a lot of variables we will not be using, so we will remove those now. We also need to add a column containing the percentage of fatalities that occured out of the total number of people aboard each flight.

```python
#Function: drop columns, drops all aside from the 5 wanted, and calculates percentage of fatalities and adds that column

def drop_cols2(overallflights):
	overallflights = overallflights[['Date', 'Operator', 'Aboard', 'Fatalities', 'Summary']]
	overallflights = overallflights.assign(FatalitiesPercent =  overallflights['Fatalities']/ overallflights['Aboard'].replace(0, 1))
	return overallflights

#Drop col 2 funct
overallflights = drop_cols2(overallflights)
```

## Exploratory Data Analysis

### Section 1 of EDA - Categorization of Different Crashes 

In this section, we categorized the crashes in three main categories: weather, human error, and mechanical error. The purpose of this section is to analyze the roots of the crash incident in the dataset, allowing us to see which type of cause is most usually responsible for incidents.


#### Graph 1 - Total Count of Crashes by Incident Category

```python

def categorize_incident(text):
    """
    Categorizes the incident into Weather, Human Error, or Mechanical Error
    based on keyword occurrences in the processed probable cause text.
    """

    # The top 30 words were categorized into the three categories below manually
    weather_keywords = {'turbulence', 'air', 'conditions', 'weather', 'bird'} # Where 'bird' was noticed after manually looking at dataset
    human_mech_incidentsor_keywords = {'crew', 'captain', 'pilot', 'attendant', 'personnel', 'operator', 'driver', 'traffic', 'passenger', 'service', 'decision', 'officer'}
    mechanical_error_keywords = {'engine', 'gear', 'maintenance', 'system', 'fuel', 'controller', 'separation', 'fire', 'pressure', 'hydraulic', 'fan', 'nose', 'power',
                                'blade', 'turbine'}
    
    words = set(text.split())  # Convert text into a set of words for quick lookup
    
    if words & weather_keywords:
        return 'Weather'
    elif words & human_mech_incidentsor_keywords:
        return 'Human error'
    elif words & mechanical_error_keywords:
        return 'Mechanical error'
    else:
        return 'Other'  # Default category if no keywords match
```

```python
# Apply function to create new category column
commercialflights['IncidentCategory'] = commercialflights['ProbableCause'].apply(categorize_incident)
```

```python
# To check if the incidents were categorized right
commercialflights[['ProbableCause', 'IncidentCategory']]
```

```python
# Count occurrences of each category
category_counts = commercialflights['IncidentCategory'].value_counts()
```

```python
# Plot bar chart
plt.figure(figsize=(10, 5))
sns.barplot(x=category_counts.index, y=category_counts.values, palette="pastel", hue=category_counts.index, legend=False)
plt.xlabel("Incident Category")
plt.ylabel("Total Count of Crashes")
plt.title("Total Count of Crashes by Incident Category")

# Show values on bars
for i, value in enumerate(category_counts.values):
    plt.text(i, value + 1, str(value), ha='center', fontsize=12)

plt.show()
```

```python
category_proportions = category_counts / category_counts.sum()

# Plot bar chart for proportions
plt.figure(figsize=(10, 5))
sns.barplot(x=category_proportions.index, y=category_proportions.values, hue=category_counts.index, palette="pastel", legend=False)
plt.ylim(0, max(category_proportions.values) + 0.05) # Because initial plot without y-axis limit had the percentage overlapping with border

# Add labels and title
plt.xlabel("Incident Category")
plt.ylabel("Proportion of Crashes")
plt.title("Proportion of Crashes by Incident Category")

# Show values on bars
for i, value in enumerate(category_proportions.values):
    plt.text(i, value + 0.01, f"{value:.2%}", ha='center', fontsize=12)

plt.show()
```

These two graphs plot the NTSB dataset. The first one illustrates the total amount of crashes (y-axis) per incident category (x-axis), whereas the second plot illustrates the proportions of crashes (y-axis) per incident category (x-axis). Results show that human error is the most likely cause of aircraft incidents, followed by weather, mechanical error, and then other causes that fall outside of the three main categories as determined by our manual categorization of the 30 most common words in the “Probable Cause” column.

#### Graph 3 - Patterns of Safety Incidents from Human and Mechanical Errors Over Time

This is a graph that shows trends over time for the NTSB dataset, specifically for incident categories of human error and mechanical error. It is plotted as a kde of a histogram with each bin being a year. A quick look shows that overall there are more human error incidents compared to mechanical ones, although the gap between the yearly frequency of each cause is decreasing as the number of crashes by either cause decreases overall.

Indeed there is not a single year with more mechnical incidents than human ones. Aditionally there seems to be a slight downward trend over time for both. Indicating while human error is more predominant, both are slowly decreasing over time.

# Ethics & Privacy

Dataset #1 comes from the National Transportation Safety Board, which is a federal agency that investigates U.S. aviation accidents. Each accident observation has a docket report linked to it which contains flight crew members’ full names and ages. As this personally identifiable information is not relevant to our analysis of airplane crashes, we did not include the data column with the links to the reports. Additionally, we removed the NTSB identification numbers for the incidents so that the reports can not be easily accessed. However, since all information on NTSB crashes is publicly available, we acknowledge that the names of flight crew members could still be accessed. Our second dataset, accessed from Kaggle, does not contain any personally identifiable information. 

An additional privacy concern regarding our datasets includes the handling of data on the number of injuries and fatalities in plane crashes, which, in combination with the date and airline operator of the crashes, could possibly be linked to certain passengers or crew members. We deem the privacy risk to be low, in part due to our choice to remove the locations of the crashes. 

A concern of possible bias we identify is the current presence of strong public opinions regarding the safety of particular U.S. airlines. Additionally, our objective analysis of plane crashes across different airlines does not intend to direct distrust or any otherwise negative publicity towards any commercial airline, but rather is concerned with investigating flight accidents from a safety standpoint.
